```python
driver_complete=pd.read_csv("/kaggle/input/formula-1-world-championship-1950-2020/drivers.csv")
driver=driver_complete[["driverId","code"]]
driver["name"]=driver_complete["forename"]+' '+driver_complete["surname"]
driver
```

```python
cons=pd.read_csv("/kaggle/input/formula-1-world-championship-1950-2020/constructors.csv")
cons=cons.drop(["url","nationality"],axis=1)
cons
```

```python
cons_res=pd.read_csv("/kaggle/input/formula-1-world-championship-1950-2020/constructor_results.csv")
cons_res.drop(["status"],axis=1)
```

```python
res=pd.read_csv("/kaggle/input/formula-1-world-championship-1950-2020/results.csv")
res=res.drop(["time","milliseconds","fastestLapTime","fastestLapSpeed","statusId","number","positionText","positionOrder","rank"],axis=1)
res=res.drop(["laps","fastestLap"],axis=1)
res.rename(columns={"position":"RacePos"})
res
```

```python
driver_stand=pd.read_csv("/kaggle/input/formula-1-world-championship-1950-2020/driver_standings.csv")
driver_stand=driver_stand.drop(["driverStandingsId","positionText"],axis=1)
driver_stand
```

```python
# Create aggregated table with custom column names
driver_stats = driver_stand.groupby('driverId').agg(
    TotalWins=('wins', 'sum'),
    TotalPointsWon=('points', 'sum')
).reset_index()
driver_stats
```

```python
races=pd.read_csv("/kaggle/input/formula-1-world-championship-1950-2020/races.csv")
races=races[races["year"]>2019]
races=races[["raceId","year","circuitId","name"]]
races
```

```python
qualif=pd.read_csv("/kaggle/input/formula-1-world-championship-1950-2020/qualifying.csv")
quali=qualif[["qualifyId","raceId","driverId","constructorId","position"]]
quali
```

```python
cont=pd.merge(quali,races,on="raceId")
#cont.sort_values("raceId")
cont=cont.rename(columns={"position":"QualiPos"})
cont
```

```python
rev=pd.merge(cont,res,on=["raceId","driverId","constructorId"])
# Replace \N with 20 in the position column
rev['position'] = rev['position'].replace('\\N', 20)
rev
```

```python
rec=pd.merge(rev,driver,on="driverId")
rec=rec.rename(columns={"name_x":"RaceName","name_y":"DriverName","position":"RacePos"})
rec=rec.drop(["circuitId","points"],axis=1)
rec
```

```python
df=rec
# Convert RacePos to numeric (DNFs/retirements become NaN)
df['RacePos'] = pd.to_numeric(df['RacePos'], errors='coerce')

# Sort by driverId, year, raceId to ensure correct order for rolling calculations
df = df.sort_values(['driverId', 'year', 'raceId'])

df['driver_last3_avg'] = df.groupby('driverId')['RacePos'].transform(
    lambda x: x.rolling(3, min_periods=1).mean().shift(1)
)

df['constructor_last3_avg'] = df.groupby('constructorId')['RacePos'].transform(
    lambda x: x.rolling(3, min_periods=1).mean().shift(1)
)

df['driver_total_races'] = df.groupby('driverId').cumcount()
df['driver_wins'] = df.groupby('driverId')['RacePos'].transform(
    lambda x: (x == 1).cumsum().shift(1).fillna(0)
)
df['driver_podiums'] = df.groupby('driverId')['RacePos'].transform(
    lambda x: (x <= 3).cumsum().shift(1).fillna(0)
)

df['driver_win_rate'] = df['driver_wins'] / (df['driver_total_races'] + 1)
```

```python
# Sort by driver and race date

df = df.sort_values(['driverId', 'year', 'raceId'])

# Last 3 races average finish position for each driver
df['driver_last3_avg'] = df.groupby('driverId')['RacePos'].transform(
    lambda x: x.rolling(3, min_periods=1).mean().shift(1)
)

# Last 3 races average for constructor
df['constructor_last3_avg'] = df.groupby('constructorId')['RacePos'].transform(
    lambda x: x.rolling(3, min_periods=1).mean().shift(1)
)
# Calculate cumulative stats (up to current race)
df['driver_total_races'] = df.groupby('driverId').cumcount()
df['driver_wins'] = df.groupby('driverId')['RacePos'].transform(
    lambda x: (x == 1).cumsum().shift(1).fillna(0)
)
df['driver_podiums'] = df.groupby('driverId')['RacePos'].transform(
    lambda x: (x <= 3).cumsum().shift(1).fillna(0)
)

# Win rate
df['driver_win_rate'] = df['driver_wins'] / (df['driver_total_races'] + 1)

df
```

```python
# Correct way to check nulls
df.isnull().sum()
```

```python
# Fill NaN with reasonable defaults
df['driver_last3_avg'].fillna(10, inplace=True)  # Middle grid position
df['constructor_last3_avg'].fillna(10, inplace=True)
df['driver_win_rate'].fillna(0, inplace=True)  # New drivers have 0 wins

# Or fill with global average
# df['driver_last3_avg'].fillna(df['RacePos'].mean(), inplace=True)
df.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score,classification_report
```

```python
features = ['grid', 'driver_last3_avg', 'constructor_last3_avg', 'driver_win_rate', 'driver_total_races']
df['top10'] = (df['RacePos'] <= 10).astype(int)
```

```python
X_train, X_test, y_train, y_test=train_test_split(df[features],df['top10'],test_size=0.2,random_state=81)
```

```python
#RandomForest
rfc=RandomForestClassifier(n_estimators=100,max_depth=10,random_state=81)
rfc.fit(X_train,y_train)
rfc_pred=rfc.predict(X_test)
```

```python
#XGBoost Classifier
xgb=XGBClassifier(n_estimators=100,max_depth=6,learning_rate=0.1)
xgb.fit(X_train,y_train)
xgb_pred=xgb.predict(X_test)
```

```python
print("Random Forest Accuracy:", accuracy_score(y_test, rfc_pred))
print("\nXGBoost Accuracy:", accuracy_score(y_test, xgb_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print("---Random Forest Classifier---\n")
print(classification_report(y_test, rfc_pred))
print(confusion_matrix(y_test, rfc_pred))
print("\n\n---XGBoost Classifier---\n")
print(classification_report(y_test, xgb_pred))
print(confusion_matrix(y_test, xgb_pred))
```

```python
print("Feature Importance\n")
feature_imp = pd.DataFrame({
    'feature': features,
    'importance': rfc.feature_importances_
}).sort_values('importance', ascending=False)
print(feature_imp)
```

```python
#Trying GridSearchCV on RandomForest for better hyperparameters
param_grid_rf={
    'n_estimators':[50,100,200],
    'max_depth':[8,10,15,None],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4],
    'max_features':['sqrt','log2'],
    'bootstrap':[True,False]
}
grid_rf=GridSearchCV(
    estimator=RandomForestClassifier(random_state=81),
    param_grid=param_grid_rf,
    cv=3,
    scoring='accuracy',
    verbose=2,
    n_jobs=1
)
grid_rf.fit(X_train,y_train)

print("Best RF params:", grid_rf.best_params_)
print("Best RF score:", grid_rf.best_score_)

best_rf=grid_rf.best_estimator_
rf_pred_tuned=best_rf.predict(X_test)
```

```python
#Trying GridSearchCV on XGBoost Classifier for better hyperparameters
param_grid_xgb={
    'n_estimators':[50,100,200],
    'max_depth':[3,5,7],
    'learning_rate':[0.01,0.1,0.3],
    'subsample':[0.8,1.0],
    'min_child_weight':[1,3,5]
}
grid_xgb=GridSearchCV(
    estimator=XGBClassifier(random_state=81),
    param_grid=param_grid_xgb, 
    cv=3,
    scoring='accuracy',
    verbose=2,
    n_jobs=1
)
grid_xgb.fit(X_train, y_train)

print("Best XGB params:", grid_xgb.best_params_)
print("Best XGB score:", grid_xgb.best_score_)

best_xgb=grid_xgb.best_estimator_
xgb_pred_tuned=best_xgb.predict(X_test)
```

```python
print("After GridSearchCV on both to find the optimal hyperparameters:\n")
print("Tuned Random Forest Accuracy:", accuracy_score(y_test, rf_pred_tuned))
print("\nTuned XGBoost Accuracy:", accuracy_score(y_test, xgb_pred_tuned))
```

```python
print("After GridSearchCV on both of the models:\n")
print("---Tuned Random Forest Classifier---\n")
print(classification_report(y_test, rf_pred_tuned))
print(confusion_matrix(y_test, rf_pred_tuned))
print("\n\n---Tuned XGBoost Classifier---\n")
print(classification_report(y_test, xgb_pred_tuned))
print(confusion_matrix(y_test, xgb_pred_tuned))
```
